```python
"""
Import require libraries for graph construction,
numerical computation and visualization.
"""
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""
Function to comoute average node degree of a graph.
Average node degree represents how many connections
each node has on average.
"""
def average_degree(G):
    degree = dict(G.degree())
    return sum(degree.values()) / len(degree)
```

##Experimental Design

To perform a computational experiment, graphs are generated
using different connection probabilities(p).

For each possibilty values:
1. A random graph is generated using NetworkX.
2. The average node degree is calculated.
3. Results are stored for comparison.

This allows observation of how connectivity changes as randomness increases.

```python
"""
Generate graphs for multiple probability values
and store experimental results.
"""

probabilities = np.linspace(0.01, 0.5, 12)
ave_degrees_results = []
for p in probabilities:
    G = nx.erdos_renyi_graph(50, p)
    ave_degrees_results.append(average_degree(G))
```

```python
"""
Visualization relationship between connection probability and average network degree.
"""
plt.figure()
plt.plot(probabilities, ave_degrees_results)
plt.xlabel("Connection Probability(p)")
plt.ylabel("Average Degree")
plt.title("Network Connectivity vs Randomness")
plt.grid(True)
plt.show()
```

##Analysis

The result show that as connection probability increases,
the average degree also increase.
This means nodes become more connected as randomness grows.

The experiment demostrates how simple probabilistic rules can lead to increasingly complex network structures.
